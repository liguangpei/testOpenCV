# 4.1线性模型的基本概念
##常用的线性模型：线性回归，岭回归，套索回归，逻辑回归和线性SVC

# 画一条直线的解析方程式

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 100)
y = 0.5*x+3
#print("x=\n{}".format(x))
#print("y=\n{}".format(y))
plt.plot(x, y, c='k')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=x*0.5+3")
plt.show()
```

# 4.1.2根据2个点确定一条直线，并计算直线方程

```python
#导入线性回归
from sklearn.linear_model import LinearRegression
#输入2个点的x
X=[[1],[4]]
y = [3, 5]

#用线性模型拟合
lr = LinearRegression()
lr.fit(X, y)

#根据x坐标预测y坐标
z = np.linspace(0, 5, 20)
z = z.reshape(-1, 1)
plt.plot(z, lr.predict(z), c='k')
plt.scatter(X, y, c='r')

plt.xlabel("x")
plt.ylabel("y")
plt.title('Straight Line')
print("直线方程:\ny=x*{:.3f}".format(lr.coef_[0]),"+{:.3f}".format(lr.intercept_))
plt.show()

```

根据3个点确定一条直线，并计算直线方程

```python
#导入线性回归
from sklearn.linear_model import LinearRegression
#输入3个点的x
X=[[1],[4],[3]]
y = [3, 5,3]

#用线性模型拟合
lr = LinearRegression()
lr.fit(X, y)

#根据x坐标预测y坐标
z = np.linspace(0, 5, 20)
z = z.reshape(-1, 1)
plt.plot(z, lr.predict(z), c='k')
plt.scatter(X, y, c='r')

plt.xlabel("x")
plt.ylabel("y")
plt.title('Straight Line')
print("直线方程:\ny=x*{:.3f}".format(lr.coef_[0]),"+{:.3f}".format(lr.intercept_))
plt.show()
```

用更多数据确定一条直线方程

```python
#导入线性回归
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
#生成回归分析的数据集
X, y = make_regression(n_samples=50, n_features=1, n_informative=1, noise=50, random_state=1)

#用线性模型拟合
lr = LinearRegression()
lr.fit(X, y)

#根据x坐标预测y坐标
z = np.linspace(-3, 3, 200)
z = z.reshape(-1, 1)
plt.plot(z, lr.predict(z), c='k')
plt.scatter(X, y, c='r')

plt.xlabel("x")
plt.ylabel("y")
plt.title('Straight Line')
print("直线方程:\ny=x*{:.3f}".format(lr.coef_[0]),"+{:.3f}".format(lr.intercept_))
plt.show()
```

# 4.2最基本的线性回归模型--线性回归
4.2.1 线性回归级别原理：找到训练数据集中y的预测值和真实值的平方差最小的时候所对应的权重w和截距b。y=w*x+b

```python
#导入线性回归
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#生成回归分析的数据集
X, y = make_regression(n_samples=100, n_features=2, n_informative=2, random_state=38)
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=8)

#用线性模型拟合
lr = LinearRegression()
lr.fit(X_train, y_train)

#根据x坐标预测y坐标
z = np.linspace(-3, 3, 200)
z = z.reshape(-1, 2)
plt.plot(z, lr.predict(z), c='k')
print(X_train.shape)
print(y_train.shape)
plt.scatter(X_train[:,0], y_train, c='r')
plt.scatter(X_train[:,1], y_train, c='b')

plt.xlabel("x")
plt.ylabel("y")
plt.title('Straight Line')
print("lr.coef_:{}".format(lr.coef_[:]))
print("lr.intercept_:{}".format(lr.intercept_))
print("直线方程:\ny=x1*{:.3f}".format(lr.coef_[0]),"+x2*{:.3f}".format(lr.coef_[0]),"+{:.2e}".format(lr.intercept_))
print("==========================================")
print("训练数据集得分:{:.2f}".format(lr.score(X_train, y_train)))
print("测试数据集得分:{:.2f}".format(lr.score(X_test, y_test)))
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
#加载糖尿病数据集
from sklearn.datasets import load_diabetes
X, y = load_diabetes().data, load_diabetes().target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8)
lr = LinearRegression()
lr.fit(X_train, y_train)
print("==========================================")
print("训练数据集得分:{:.2f}".format(lr.score(X_train, y_train)))
print("测试数据集得分:{:.2f}".format(lr.score(X_test, y_test)))
```

# 4.3使用L2正则化的线性模型——岭回归

岭回归实际上是一种改良的最小二乘法，也是回归分析中常用的线性模型

```python
#导入岭回归
from sklearn.linear_model import Ridge
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
#生成回归分析的数据集
X, y = load_boston().data, load_boston().target
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=8)
ridge = Ridge()
ridge.fit(X_train, y_train)
print("==========================================")
print("训练数据集得分:{:.2f}".format(ridge.score(X_train, y_train)))
print("测试数据集得分:{:.2f}".format(ridge.score(X_test, y_test)))
```

#4.3.2岭回归参数调节

```python
from sklearn.linear_model import LinearRegression
#加载糖尿病数据集
from sklearn.datasets import load_diabetes
X, y = load_diabetes().data, load_diabetes().target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8)
ridge10 =Ridge(alpha=10)#通过提升alpha值，可以降低过拟合
ridge10.fit(X_train, y_train)
print("==========================================")
print("训练数据集得分:{:.2f}".format(ridge10.score(X_train, y_train)))
print("测试数据集得分:{:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
#如果alpha值过低，则得到的结果就会非常接近线性回归
from sklearn.linear_model import LinearRegression
#加载糖尿病数据集
from sklearn.datasets import load_diabetes
X, y = load_diabetes().data, load_diabetes().target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8)
ridge01 =Ridge(alpha=0.01)#通过提升alpha值，可以降低过拟合
ridge01.fit(X_train, y_train)
print("==========================================")
print("训练数据集得分:{:.2f}".format(ridge01.score(X_train, y_train)))
print("测试数据集得分:{:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
#alpha值高，则对模型的限制严格coef_属性值越小。
#alpha值越低，对模型的限制越松coef_属性值越大
plt.plot(ridge.coef_, 's', label='Ridge alpha=1')
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")
plt.plot(ridge01.coef_, '^', label="Ridge alpha=01")
#绘制线性回归的系数作为对比
plt.plot(lr.coef_, 'o', label="linear regression")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.hlines(0,0,len(lr.coef_))
plt.legend()
```
